```python
#Importing libraries
import requests
from pandas import DataFrame
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Importing dataset
dfscore=pd.read_csv(r"D:\dataset.csv")
```

```python
#Obtaining a unique list of journals
test = []
for title in dfscore.publication_CR:
        test.append(title)
    
```

```python
#the length of unique journal names
journals = set(test)
len(journals)
```

```python
#Checking the website for response code
url = 'https://www.scimagojr.com/journalrank.php?page='
response = requests.get(url)
print(response)
```

```python
#Looping through the website url for going through multiple pages.
#Also, scraping the table to obtain a list of lists containing journal and its attributes.
buffer = []

for i in range(1,641):
    url = 'https://www.scimagojr.com/journalrank.php?page=' + str(i)
    response = requests.get(url)
    

    soup = BeautifulSoup(response.content, "lxml")


    for mytable in soup.find_all('tbody'):
    
                for trs in mytable.find_all('tr'):
                    tds = trs.find_all('td')
                    row = [elem.text.strip() for elem in tds]
                    buffer.append(row)
```

```python
buffer
```

```python
len(buffer)
```

```python
final_list = []
#Obtaining H-Index and Citations with respective journal names
for i in range(len(buffer)):
    final_list.append(buffer[i][1]) 
    final_list.append(buffer[i][4])
    final_list.append(buffer[i][8])
```

```python
final_list
```

```python
#Transforming data for bringing it to a useful format.
i=0
new_list=[]
while i<len(final_list):
    new_list.append(final_list[i:i+3])
    i+=3
```

```python
new_list
```

```python
df = DataFrame(new_list, columns = ['publication_CR','H-Index','Total_Citation'])
```
