```python
#importing libraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
#Checking the website for response code
url = 'http://eigenfactor.org/projects/journalRank/journalsearch.php'
response = requests.get(url)
print(response)
```

```python
#reading scores csv file
dfscore=pd.read_csv(r"C:\Users\prach\Desktop\SCORE_csv.csv")

#created a set of journal names 
test = []
for title in dfscore.publication_CR:
        test.append(title)
journals = set(test)

#created a list of the journal names
finalist=list(journals)
finalist.sort()
#finalist
```

```python
#printing teh orginal dataset
dfscore
```

```python


varlist1 = []
okay = []
#coded the for loop which loops through the length of the unique journal names
for i in range(len(finalist)):
    url = 'http://eigenfactor.org/projects/journalRank/rankings.php?search={}&searchby=exact&year=2015&orderby=Eigenfactor'.format(finalist[i])

    #storing the contents of the url in the varaible data 
    data = requests.get(url)
    #storing the scrapped contents in soup 
    soup = BeautifulSoup(data.content, 'html.parser')
    #scraped the div class ef which contains the Eigen value for the specific journal name
    mydivs = soup.findAll("div", {"class": "EF"})
    mydivs
    #we convert the resultset to list named=listofef
    listofef =[]
    for x in mydivs:
        listofef.append(str(x))
        
    listofef
    final= str(listofef)
    
    #extracts EF value 

    eflist=[]

    #Regular expression, to extract the float eigen value from final var
    for line in re.findall("[\d]*[.][\d]+", final):
        #print(line)
        eflist.append(line)
        #print(eflist)
        
        
     #condotion for the journal names that were not found,
     #if the list length turns out to be 0, we continue and the for loop continues   
    if(len(eflist)==0):
        continue
            
    else:
        varlist =[]
        varlist = eflist[0]
        varlist1.append(varlist)
        okay.append(finalist[i])
        #print(okay)
        print("Journal Name:",finalist[i],"| EigenFactor:",varlist)
```

```python
#now we combine the journal names and egiven values in a new list named new_lst
new_lst = [list(x) for x in zip(okay, varlist1)]
new_lst
```

```python
#create a dataframe which contains journal name and eigen value
df = pd.DataFrame(new_lst,columns = ['publication_CR','Eigen Factor'])
df
```

```python
#using merge function of pandas we merged the eigen value on journal name using left join 

final = pd.merge(dfscore, df, on='publication_CR', how='left')
```
